###   Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word.
2. Remove Punctuation.
3. Remove Stopwords.

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Instantiate the lemmatizer
# YOUR CODE HERE!
lemmatizer = WordNetLemmatizer()

# Create a list of stopwords
# YOUR CODE HERE!
stop = stopwords.words('english') #this is a list

# Expand the default stopwords list if necessary #run the whole thing and come back to this step because we don't know yet 
# YOUR CODE HERE!
stop.append('chars')
```

```python
# Complete the tokenizer function
def tokenizer(text):
   
    # Remove the punctuation from text
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', text)
   
    # Create a tokenized list of the words
    words = word_tokenize(re_clean)
    
    # Lemmatize words into root words
    lem = [lemmatizer.lemmatize(word) for word in words]
   
    # Convert the words to lowercase and cleaning up stop words
    tokens = [word.lower() for word in lem if word.lower() not in stop]
    
    return tokens
```

```python
# Create a new tokens column for Bitcoin
# YOUR CODE HERE!
bitcoin_df['tokens'] = bitcoin_df.text.apply(tokenizer) #apply a function to every row in the dataframe 
bitcoin_df.head()
```

```python
# Create a new tokens column for Ethereum
# YOUR CODE HERE!
ethereum_df['tokens'] = ethereum_df.text.apply(tokenizer)
ethereum_df.head()
```

### NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
# YOUR CODE HERE!
bigram_counts_bitcoin = Counter(ngrams(tokenizer(bitcoin_df['text'].str.cat()), n=2))
print(dict(bigram_counts_bitcoin))
```

```python
# Generate the Ethereum N-grams where N=2
# YOUR CODE HERE!
bigram_counts_ethereum = Counter(ngrams(tokenizer(ethereum_df['text'].str.cat()), n=2))
print(dict(bigram_counts_ethereum))
```

```python
# Function token_count generates the top 10 words for a given coin
def token_count(tokens, N=3):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Use token_count to get the top 10 words for Bitcoin
# YOUR CODE HERE!
token_count(tokenizer(bitcoin_df['text'].str.cat()), 10)
```

```python
# Use token_count to get the top 10 words for Ethereum
# YOUR CODE HERE!
token_count(tokenizer(ethereum_df['text'].str.cat()), 10)
```

### Bitcoin NER

```python
# Concatenate all of the Bitcoin text together
# YOUR CODE HERE!
bitcoin_text = bitcoin_df['text'].str.cat()
bitcoin_text
```

```python
# Run the NER processor on all of the text
# YOUR CODE HERE!
# Set article to be analyzed with spaCy
bitcoin_NER = nlp(bitcoin_text)

# Render NER visualization with displacy to determine entities for extraction
displacy.render(bitcoin_NER, style='ent')

# Add a title to the document
# YOUR CODE HERE!
```

```python
# List all Entities
# YOUR CODE HERE!
def show_ents(doc):
    if doc.ents:
        for ent in doc.ents:
            print(ent.text+' - '+ent.label_+' - '+str(spacy.explain(ent.label_)))
    else:
        print('No named entities found.')
            
        
show_ents(bitcoin_NER)

# for ent in bitcoin_text.ents:
#     print(ent.text, ent.label_)
```
